# Nomalized Least-Mean-Squares (NLMS)

O NLMS é uma variante do algoritmo LMS criada para resolver o problema de sensibilidade do LMS a valores muito grandes na entrada, fazendo com que o algoritmo "exploda" e apresente problemas de estabilidade. A ideia do NLMS é utilizar um coeficiente de aprendizado normalizado $\eta(k)$ ao invés do $\mu$ constante do LMS. 
O valor de $\eta(k)$ é calculado a cada iteração com base na potência do vetor de entrada conforme a seguinte relação:

\begin{align}
    \eta(k) = \frac{\mu}{\epsilon + ||X(k)||^2}
\end{align}

em que $||X(k)||^2$ é a norma do vetor de entrada e $\epsilon$ é o chamado fator de regularização, uma pequena constante positiva utilizada para garantir a estabilidade computacional casp o vetor de entrada tenda a zero.

# Previsão com NLMS

O seguinte exemplo apresenta o uso de filtragem adaptativa para previsão de sinais.

```python
# -*- coding: utf-8 -*-
"""
Created on Tue May 28 23:15:32 2019

@author: Victor
"""

import numpy as np
import matplotlib.pylab as plt
import padasip as pa
import pandas as pd
from padasip.misc.error_evaluation import get_mean_error


#================================================================================
# Definição dos parâmetros do filtro
n_dia = 96
N = 96
eps = 1e-6
mu = 0.5
K = 50
#================================================================================
#Carrega os dados que serão utilizados no treinamento
dataframe = pd.read_csv('data.csv',header=None,names=['data', 'hora', 'leitura'], parse_dates=[['data', 'hora']])
split = len(dataframe) - n_dia

df_train = dataframe.iloc[:split]
df_test = dataframe.iloc[split:].reset_index(drop=True)
df_prediction = dataframe.iloc[split:].reset_index(drop=True)


d_train = df_train.get('leitura').values[:]
d_test = df_test.get('leitura').values[:]

#================================================================================
#Cria o objeto da clase FilterNLMS 
filtro = pa.filters.FilterNLMS(N, mu, eps, 'zeros')
#================================================================================
#identificação
prefixedInput = np.insert(d_train, 0 , np.zeros(N))
saida_filtro = np.zeros(len(d_train))
erro = np.zeros((K,len(d_train)))
mse = np.zeros(K)
for j in range(K):
    for i in range(len(d_train)):
        regressor = prefixedInput[i:i+N]
        saida_filtro[i], erro[j][i] = filtro.adapt(d_train[i], regressor)
    mse[j] = np.average(erro[j])
   
previsao = np.zeros(len(d_test))
regressor = d_train[-1:len(d_train)-(N+1):-1]
for i in range(len(d_test)):
   previsao[i] = filtro.predict(regressor)
   #Como estamos fazendo uma previsão, devemos atualizar o regressor 
   #com o último valor previsto
   regressor = np.insert(regressor[:-1], 0, previsao[i])

df_prediction.update(pd.DataFrame({'leitura' : previsao}))
df_prediction.rename(columns={'leitura' : 'previsao'}, inplace=True)
```

A seguir é apresentada a comparação entre os sinais previstos e os verificados.

```python
ax = df_test.plot(x='data_hora', y='leitura', figsize=(15,5))
df_prediction.plot(x='data_hora', y='previsao', ls='--', ax=ax, color='red', grid=True)
ax.legend(['Verificado','Previsão'])
plt.xlabel('Data-Hora');plt.ylabel('Corrente (A)')
```

```python
Aqui podemos ver a evolução do MSE ao longo das iterações.
```

```python
plt.figure(figsize=(15,5))
plt.plot(10*np.log10(mse**2),"r", label='MSE [dB]');plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xlabel('Número de iterações (K)');plt.ylabel('MSE [dB]')
```
